```python
from tkinter import *


def process():
    temperature = float(e1.get())
    # 다음 코드 한 줄이 화씨를 섭씨로 바꾸는 코드입니다. 
    mytemp = (temperature-32)*5/9
    e2.insert(0, str(mytemp))


window  = Tk()

# 파이썬 프로그램을 실행한 후 창이 나타나면 창 크기를 크게 해 보세요. 
# 아래 코드는 교재에 없는 내용으로 추가해야 합니다. 
window.title("섭씨 <--> 화씨 변환 프로그램")


l1 = Label(window , text="화씨")
l2 = Label(window, text="섭씨")
l1.grid(row=0, column=0)
l2.grid(row=1, column=0)


e1 = Entry(window)
e2 = Entry(window)
e1.grid(row=0, column=1)
e2.grid(row=1, column=1)

# b1 (“화씨 -> 섭씨”) 버튼을 누르면 위에 있는 process 함수가 실행됨. 
b1 = Button(window, text="화씨->섭씨", command=process) 

# b2 (“섭씨 -> 화씨”) 버튼을 누르면 위에 있는 할 일이 지정되어 있지 않음. 
b2 = Button(window, text="섭씨->화씨")
b1.grid(row=2, column=0)
b2.grid(row=2, column=1)


window.mainloop()
```
